```python
from kmds.utils.config_reader import get_config
from minio import Minio
```

```python
s3_cfg = get_config("./play_minio_cfg.yaml")
```

```python
HOST_URL = s3_cfg["HOST_URL"]

ACCESS_KEY = s3_cfg["ACCESS_KEY"]
SECRET_KEY = s3_cfg["SECRET_KEY"]
BUCKET_NAME = s3_cfg["BUCKET_NAME"]
```

```python
client = Minio(endpoint=HOST_URL, access_key=ACCESS_KEY, secret_key=SECRET_KEY, secure=True)
```

```python
import pandas as pd
from io import StringIO
object_name = "incident_event_log_02.csv"
obj = client.get_object(BUCKET_NAME, object_name)
data = obj.data
```

```python
s = str(data,'utf-8')
data = StringIO(s)
df = pd.read_csv(data)
```
