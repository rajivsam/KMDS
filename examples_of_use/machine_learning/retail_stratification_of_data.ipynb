```python
 from sklearn.cluster import HDBSCAN
import pandas as pd
```

```python
fp = "../../kmds/examples/retail_transformed_q1_sales_data.parquet"
df = pd.read_parquet(fp)
df
```

```python
hdb = HDBSCAN(min_samples=5)
hdb.fit(df)
```

```python
 hdb.labels_
```

```python
df["cluster"] = hdb.labels_
```

```python
df["cluster"].value_counts()
```
